```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_circles
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
X,y=make_circles(n_samples=750,factor=0.3,noise=0.1)
```

```python
X
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
```

```python
from sklearn.cluster import DBSCAN
```

```python
dbscan=DBSCAN(eps=0.10)
```

```python
#-1 is considered as an outlier
dbscan.fit_predict(X)
```

```python
dbscan.labels_
```

```python
total_labels=set(dbscan.labels_)
len(total_labels)
```

```python
#total labels after removing noise 
total_labels=total_labels-(1 if -1 in dbscan.labels_ else 0)
total_labels
```

```python
#now lets see the outliers wrt to labels
plt.scatter(X[:,0],X[:,1],c=dbscan.labels_)
```
