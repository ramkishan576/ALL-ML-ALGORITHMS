```python
import seaborn as sns
```

```python
#Basic plotting with seaborn
tips_df=sns.load_dataset("tips")
tips_df
```

```python
type(tips_df)
```

```python
#Creating a scatter plot
```

```python
import matplotlib.pyplot as plt
sns.scatterplot(x="total_bill",y="tip",hue="time",data=tips_df)
```

```python
tips_df.info()
```

```python
##Line Plot
sns.lineplot(x="size",y="total_bill",data=tips_df)
plt.title("SIZE vs TOTAL BILL")
plt.show()
```

```python
tips_df.head()
```

```python
##categorical plots
##bar plot
sns.barplot(x="day",y="total_bill",data=tips_df)
plt.title("BAR PLOT OF TOTAL BILL BY DAY")
plt.show()
```

```python
##categorical plots
##bar plot
sns.barplot(x="smoker",y="total_bill",data=tips_df)
plt.title("BAR PLOT OF TOTAL BILL BY SMOKER")
plt.xlabel("SMOKER")
plt.ylabel("TOTAL BILL")
plt.show()
```

```python
##BOX PLOT
sns.boxplot(x="day",y="total_bill",data=tips_df)
plt.show()
```

```python
##VIOLIN PLOT
sns.violinplot(x="day",y="total_bill",data=tips_df)
plt.show()
```

```python
##HISTOGRAMS
sns.histplot(x="total_bill",bins=7,data=tips_df,kde=True)
plt.show()
```

```python
sns.kdeplot(x="total_bill",data=tips_df,fill=True)
```

```python
##Pairplot
#gives the relation between each and every attribute with other attribute
#only foe numerical variables and categorical
sns.pairplot(tips_df)
```

```python
#HEAT MAP
## realted to correlation 
corr=tips_df[["total_bill","tip",'size']].corr()
corr
```

```python
sns.heatmap(corr,cmap="coolwarm",annot=True)
```
