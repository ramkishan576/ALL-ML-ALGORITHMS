```python
dataset=load_diabetes()
```

```python
#Creating dataFrame for the dataset
```

```python
import pandas as pd
```

```python
X=pd.DataFrame(dataset["data"],columns=dataset["feature_names"])
```

```python
X.head()
```

```python
y=pd.DataFrame(dataset["target"])
```

```python
y.head()
```

```python
#Train Test Split
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=10)
```

```python
#Once try to check Correletion
X_train.corr()
```

```python
import seaborn as sns
sns.heatmap(X_train.corr(),annot=True)
```

```python
#Applying DecisiontreeRegressor
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor=DecisionTreeRegressor()
regressor.fit(X_train,y_train)
```

The above one leads to spliting the tree,Completely to its Deapth that leads to overfitting condition,to avoid that we are going to do Hyper parameter turning

```python
#Hyper parameter tuning
```

```python
param_grid={
    "criterion":["squared_error","friedman_mse","absolute_error"],
    "splitter":["best","random"],
    "max_depth":[1,2,3,5,7,10,15,25],
    "max_features":["auto","sqrt","log2"]
}
```

```python
from sklearn.model_selection import GridSearchCV
regressor=DecisionTreeRegressor()
```

```python
grid=GridSearchCV(regressor,param_grid=param_grid,cv=6,scoring="neg_mean_squared_error")
```

```python
import warnings
warnings.filterwarnings("ignore")
grid.fit(X_train,y_train)
```

```python
grid.best_estimator_
```

```python
grid.best_params_
```

```python
y_pred=grid.predict(X_test)
```

perfomance metrices

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
print(mean_squared_error(y_test,y_pred))
```

to Check the decision Tree

```python
grid.best_params_
```

```python
selected_model=DecisionTreeRegressor(criterion="squared_error",max_depth=3,max_features="log2",splitter="best")
```

```python
selected_model.fit(X_train,y_train)
```

Visualize The Best Decision Tree

```python
from sklearn import tree
import matplotlib.pyplot as plt
plt.figure(figsize=(15,10))
tree.plot_tree(selected_model,filled=True)
plt.show()
```
