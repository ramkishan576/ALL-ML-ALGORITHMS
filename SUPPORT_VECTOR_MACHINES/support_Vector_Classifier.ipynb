```python
#lets create synthetic datapoints
from sklearn.datasets import make_classification
```

```python
X,y=make_classification(n_samples=1000,n_features=2,n_classes=2,n_clusters_per_class=1,n_redundant=0)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=10)
```

```python
''' 
until now we have we have applied support vector classifier(i.e linear kernel)
now lets change the dataset(into non linear form) and apply some other kernels and see how they chnage
'''
```

```python
from sklearn.datasets import make_classification
X,y=make_classification(n_samples=1000,n_classes=2,n_redundant=0,n_informative=2,n_clusters_per_class=1,class_sep=2,flip_y=0)
```

```python
import seaborn as sns
sns.scatterplot(x=pd.DataFrame(X)[0],y=pd.DataFrame(X)[1],hue=y)
#here we can see dataset is following non linear relationship
```

```python
#now lets apply some other kernels to improve accuracy,if we apply svc it does not work good with this data
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=10)
```

```python
from sklearn.svm import SVC
poly=SVC(kernel="poly")
```

```python
poly.fit(X_train,y_train)
```

```python
#prediction
y_pred=poly.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))

#here we can observe the increase in the accuracy of the model after applying polynomila kernel rather than linear kernel
```

```python
#lets apply rbf kernel and see how the accuracy going to change
rbf=SVC(kernel="rbf")
```

```python
rbf.fit(X_train,y_train)
```

```python
y_pred=rbf.predict(X_test)
```

```python
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```

```python
rbf.intercept_
```

```python
#obviously we can coeffient for only linear kernal
rbf.coef_
```

```python
#here we can see increae in the accuracy with applying the rbf kernel
```

```python
#Hyperparameter Tuning with SVC using GridSearchCv
```

```python
from sklearn.model_selection import GridSearchCV
param_grid={
    'kernel': ['linear', 'rbf', 'poly'],
    'C': [0.1, 1, 10, 100, 100],
    'gamma': [0.1, 1, 10, 100, 100]
}
```

```python
grid=GridSearchCV(SVC(),param_grid=param_grid,refit=True,verbose=3,cv=5)
```

```python
grid.fit(X_train,y_train)
```

```python
y_pred=grid.predict(X_test)
```

```python
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```
