```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv("Algerian_forest_fires_cleaned_data.csv")
```

```python
df.head()
```

```python
#DROP day,month,year
df.drop(["day","month","year"],axis=1)
```

```python
df.head()
```

```python
##Encoding  of the categories in classes
```

```python
df["Classes"].value_counts()
```

```python
df["Classes"]=np.where(df["Classes"].str.contains("not fire"),0,1)
```

```python
#Categories in classes 
df["Classes"].value_counts()
```

```python
## plot dessity plot for all features
import seaborn as sns
#plt.style.use("seaborn")
sns.set_style("darkgrid")
df.hist(bins=50,figsize=(20,15))
plt.show()
```

```python
##percentage for pie chart
df["Classes"].value_counts()
```

```python
df["Classes"].value_counts(normalize=True)*100
```

```python
percentage=df["Classes"].value_counts(normalize=True)*100
```

```python
##plotting pie chart
classlabels=["Fire","Not fire"]
plt.figure(figsize=(12,7))
plt.pie(percentage,labels=classlabels,autopct="%1.1f%%")
plt.title("Pie Chart Of Classes")
plt.show()
```

```python
##CORRELATION
df.corr()
```

```python
plt.figure(figsize=(10,5))
sns.heatmap(df.corr(),annot=True,fmt=".2f",cmap='coolwarm')
```

```python
''''  
annot=True: This argument adds the correlation coefficient values in each cell of the heatmap.
fmt=".2f": This specifies the format of the annotations. In this case, it formats the floating-point numbers to two decimal places.
cmap='coolwarm': This sets the color map of the heatmap. You can choose any valid Matplotlib colormap, such as 'viridis', 'plasma', 'inferno', etc.
cbar=True: This argument (which is True by default) displays the color bar next to the heatmap, indicating the scale of correlation values.

'''
```

```python
#Box plots to check outliers
sns.boxplot(df["FWI"],color="green")
```
