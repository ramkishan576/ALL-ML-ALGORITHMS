```python
''' 
The dataset is converted into two sets based on Region 
from 122th index we can make a new column on the region
1) Bejaia Region dataset
2)sidi-Bel Abbes Region datset

add new column with region

'''
```

```python
df=dataset
```

```python
df.loc[:122,"Region"]=0               
df.loc[122:,"Region"]=1                                  
```

```python
df.info()
```

```python
df[["Region"]]=df[["Region"]].astype(int)
```

```python
## Removing the null values 
```

```python
df=df.dropna().reset_index(drop=True)
```

```python
df.head()
```

```python
df.shape
```

```python
df.iloc[[122]]
```

```python
df=df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
#fix the spaces in the column names
df.columns=df.columns.str.strip()
df.columns
```

```python
df.head()
```

```python
df.columns
```

```python
#changing the required columns as integer data type
df[["month","day","year","Temperature","RH","Ws"]]=df[["month","day","year","Temperature","RH","Ws"]].astype(int)
```

```python
df.info()
```

```python
#Changing the other columns to float datatype
```

```python
objects=[features for features in df.columns if df[features].dtype=="O"]
```

```python
objects
```

```python
for column in objects:
    if column!="Classes":
        df[column] = df[column].astype(float)
```

```python
df.info()
```

```python
for column in df.columns:
    if df[column].dtype=="int64":
        df[column]=df[column].astype(np.int32)
    elif df[column].dtype== "float64":
        df[column]=df[column].astype(np.float32) 
```
